## Variables y Tipados

### En Rust, todo es inmutable por defecto. Si quieres cambiar algo, tienes que pedir permiso con la palabra "mut"

```rust
// En una celda de Jupyter con el kernel evcxr
fn main() {
    // Inmutable por defecto (si intentas x = 6, fallará)
    let x = 5; 
    
    // Mutable (debes ser explícito)
    let mut y = 10;
    y = y + 5;
    
    // Tipado explícito (aunque Rust es muy inteligente deduciendo)
    let puntuacion: i32 = 100;
    
    println!("X: {}, Y: {}, Puntos: {}", x, y, puntuacion);
}
```

## Tipos de datos 

### Rust añade tipos escalares muy específicos (i8 hasta u128) y conceptos como las Tuplas y los Arrays de tamaño fijo.

```rust
// Tupla: varios tipos en uno
let persona: (&str, i32) = ("Ramon", 30);

// Array: tamaño fijo, todos del mismo tipo
let numeros: [i32; 3] = [10, 20, 30];

println!("Nombre: {}, Primer número: {}", persona.0, numeros[0]);
```

## Control de Flujo: Ciclos y Condicionales

### En Rust, el if es una expresión, lo que significa que devuelve un valor.

```rust
// El 'if' como asignación
let condicion = true;
let resultado = if condicion { 10 } else { 0 };

// Ciclos potentes
for n in 1..5 { // Rango del 1 al 4
    println!("Iteración: {}", n);
}

// El ciclo infinito controlado
let mut c = 0;
loop {
    c += 1;
    if c == 10 { break; }
}
```

## Funciones y Retornos

### En Rust, la última línea de una función sin punto y coma es el valor que se devuelve. Es elegante y reduce el ruido visual.

```rust
fn cuadrado(n: i32) -> i32 {
    n * n 
}
```
